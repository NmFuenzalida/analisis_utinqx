```python
# Crear índice de complejidad
# Riesgo: A=4, B=3, C=2, D=1
# Dependencia: 1=3, 2=2, 3=1
# Complejidad = Riesgo + Dependencia (máximo 7, mínimo 2)

def calcular_complejidad(categoria):
    if pd.isna(categoria):
        return None
    
    letra = categoria[0]  # Primera letra (A, B, C, D)
    numero = categoria[1]  # Segundo carácter (1, 2, 3)
    
    # Puntaje de riesgo
    riesgo = {'A': 4, 'B': 3, 'C': 2, 'D': 1}
    # Puntaje de dependencia
    dependencia = {'1': 3, '2': 2, '3': 1}
    
    return riesgo.get(letra, 0) + dependencia.get(numero, 0)

# Aplicar a UTINQX 2024 y 2025
utinqx_2024['COMPLEJIDAD'] = utinqx_2024['CATEGORIA'].apply(calcular_complejidad)
utinqx_2025['COMPLEJIDAD'] = utinqx_2025['CATEGORIA'].apply(calcular_complejidad)

# Verificar
print("Ejemplos de complejidad calculada:")
print(utinqx_2025[['CATEGORIA', 'COMPLEJIDAD']].drop_duplicates().sort_values('COMPLEJIDAD', ascending=False))
```

```python
# Calcular complejidad promedio por mes
complejidad_2024 = utinqx_2024.groupby('MES')['COMPLEJIDAD'].mean()
complejidad_2025 = utinqx_2025.groupby('MES')['COMPLEJIDAD'].mean()

# Gráfico comparativo
plt.figure(figsize=(10, 5))
plt.plot(complejidad_2024.index, complejidad_2024.values, marker='o', linewidth=2, label='2024')
plt.plot(complejidad_2025.index, complejidad_2025.values, marker='s', linewidth=2, label='2025')

plt.title('Complejidad Promedio de Pacientes UTINQX: 2024 vs 2025')
plt.xlabel('Mes')
plt.ylabel('Índice de Complejidad (mayor = más complejo)')
plt.xticks(range(1, 13), ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                          'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic'])
plt.legend()
plt.grid(True)
plt.ylim(5, 7)  # Ajustar escala para ver mejor las diferencias
plt.show()

# Promedios generales
print(f"\nComplejidad promedio 2024: {utinqx_2024['COMPLEJIDAD'].mean():.2f}")
print(f"Complejidad promedio 2025: {utinqx_2025['COMPLEJIDAD'].mean():.2f}")
```

```python
# Tabla comparativa de complejidad por mes
comparacion = pd.DataFrame({
    '2024': complejidad_2024.round(2),
    '2025': complejidad_2025.round(2),
    'Diferencia': (complejidad_2025 - complejidad_2024).round(2)
})

comparacion.index = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 
                     'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

print("Complejidad promedio por mes:")
print(comparacion)

print(f"\n--- RESUMEN ---")
print(f"Promedio anual 2024: {utinqx_2024['COMPLEJIDAD'].mean():.2f}")
print(f"Promedio anual 2025: {utinqx_2025['COMPLEJIDAD'].mean():.2f}")
print(f"Aumento: +{(utinqx_2025['COMPLEJIDAD'].mean() - utinqx_2024['COMPLEJIDAD'].mean()):.2f} puntos")
```

```python
# Primero agregar complejidad a UTIQX
utiqx_2024['COMPLEJIDAD'] = utiqx_2024['CATEGORIA'].apply(calcular_complejidad)
utiqx_2025['COMPLEJIDAD'] = utiqx_2025['CATEGORIA'].apply(calcular_complejidad)

# Comparar complejidad promedio entre unidades (2025)
print("=== COMPLEJIDAD PROMEDIO 2025 ===")
print(f"UTINQX: {utinqx_2025['COMPLEJIDAD'].mean():.2f}")
print(f"UTIQX:  {utiqx_2025['COMPLEJIDAD'].mean():.2f}")

# Calcular carga por enfermero
# UTINQX: 1 enfermero, UTIQX: 3 enfermeros
carga_utinqx = utinqx_2025['COMPLEJIDAD'].sum() / 1
carga_utiqx = utiqx_2025['COMPLEJIDAD'].sum() / 3

print(f"\n=== CARGA DE COMPLEJIDAD TOTAL POR ENFERMERO (2025) ===")
print(f"UTINQX: {carga_utinqx:.0f} puntos de complejidad / 1 enfermero")
print(f"UTIQX:  {carga_utiqx:.0f} puntos de complejidad / 3 enfermeros")
print(f"\nDiferencia: El enfermero de UTINQX maneja {carga_utinqx/carga_utiqx:.1f}x menos carga")
```

```python
# Crear columna MES para UTIQX (nos faltaba)
utiqx_2024['MES'] = utiqx_2024['FECHA_CATEGORIZACION'].dt.month
utiqx_2025['MES'] = utiqx_2025['FECHA_CATEGORIZACION'].dt.month

# Ahora sí, filtrar desde Abril
utinqx_2025_abril = utinqx_2025[utinqx_2025['MES'] >= 4]
utiqx_2025_abril = utiqx_2025[utiqx_2025['MES'] >= 4]
utinqx_2024_abril = utinqx_2024[utinqx_2024['MES'] >= 4]

# Carga por enfermero desde Abril
carga_utinqx_2024_abril = utinqx_2024_abril['COMPLEJIDAD'].sum() / 1
carga_utinqx_2025_abril = utinqx_2025_abril['COMPLEJIDAD'].sum() / 1
carga_utiqx_2025_abril = utiqx_2025_abril['COMPLEJIDAD'].sum() / 3

print("=== CARGA POR ENFERMERO (ABRIL - DICIEMBRE) ===\n")
print(f"UTINQX 2024: {carga_utinqx_2024_abril:,.0f} puntos")
print(f"UTINQX 2025: {carga_utinqx_2025_abril:,.0f} puntos")
print(f"UTIQX 2025:  {carga_utiqx_2025_abril:,.0f} puntos")

aumento = ((carga_utinqx_2025_abril - carga_utinqx_2024_abril) / carga_utinqx_2024_abril) * 100
diferencia_vs_utiqx = carga_utinqx_2025_abril / carga_utiqx_2025_abril

print(f"\n Aumento UTINQX 2024→2025: +{aumento:.1f}%")
print(f" UTINQX maneja {diferencia_vs_utiqx:.1f}x más carga que UTIQX (desde Abril)")
```

```python
# Definir categorías de alto riesgo (A y B = Máximo y Alto Riesgo)
def es_alto_riesgo(categoria):
    if pd.isna(categoria):
        return False
    return categoria[0] in ['A', 'B']  # Si empieza con A o B

# Aplicar a todos los datasets
utinqx_2024['ALTO_RIESGO'] = utinqx_2024['CATEGORIA'].apply(es_alto_riesgo)
utinqx_2025['ALTO_RIESGO'] = utinqx_2025['CATEGORIA'].apply(es_alto_riesgo)
utiqx_2024['ALTO_RIESGO'] = utiqx_2024['CATEGORIA'].apply(es_alto_riesgo)
utiqx_2025['ALTO_RIESGO'] = utiqx_2025['CATEGORIA'].apply(es_alto_riesgo)

# Calcular porcentajes
pct_utinqx_2024 = utinqx_2024['ALTO_RIESGO'].mean() * 100
pct_utinqx_2025 = utinqx_2025['ALTO_RIESGO'].mean() * 100
pct_utiqx_2024 = utiqx_2024['ALTO_RIESGO'].mean() * 100
pct_utiqx_2025 = utiqx_2025['ALTO_RIESGO'].mean() * 100

print("=== PORCENTAJE DE PACIENTES DE ALTO RIESGO (Categorías A y B) ===\n")
print(f"UTINQX 2024: {pct_utinqx_2024:.1f}%")
print(f"UTINQX 2025: {pct_utinqx_2025:.1f}%")
print(f"UTIQX 2024:  {pct_utiqx_2024:.1f}%")
print(f"UTIQX 2025:  {pct_utiqx_2025:.1f}%")
```
